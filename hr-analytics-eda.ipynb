```python
plt.figure(figsize=(12, 5))
sns.histplot(dataset.DeptID.values)
plt.show()
```

##### Histogram göstərir ki, işçilərin böyük hissəsi DeptID = 5 olan departamentdə cəmlənib. Digər departamentlərdə işçi sayı xeyli azdır. Bu isə

##### təşkilatda əsas işçi yükünün bir departament üzərində toplandığını göstərir

```python
plt.figure(figsize=(8, 8))
dataset['Sex'].value_counts().plot(kind='pie',autopct='%1.1f%%');
plt.title('İşçilərin gender bölgüsü');
plt.show()
```

##### Bu qrafik işçilərin gender üzrə paylanmasını göstərir. Məlum olur ki, əməkdaşların təxminən 56.6%-i qadın, 43.4%-i isə kişidir. Gender bölgüsü 

##### ümumilikdə balanslıdır, lakin qadın əməkdaşların payı bir qədər çoxdur.

```python
plt.figure(figsize = [20, 5])
ordered = dataset['TermReason'].value_counts().index[1:]
sns.countplot(data=dataset, x= 'TermReason', order = ordered)
plt.title('İşdən ayrılma səbəbləri')
plt.xticks(rotation=45)

plt.show()
```

##### Ayrılma səbəblərinin paylanması göstərir ki, işçilər daha çox karyera dəyişikliyi, maaş və iş məmnuniyyəti ilə bağlı səbəblərə görə şirkəti tərk edir.

```python
plt.figure(figsize = [15, 5])
Ord = dataset['ManagerName'].value_counts().index
sns.countplot(data= dataset, x='ManagerName', hue ='PerformanceScore', order=Ord)
plt.xticks(rotation=45)
plt.legend(loc='upper right')
plt.title('Menecerə üzrə performans göstəriciləri')
plt.legend(
    title="Performans",
    labels=[
        "Gözləntiləri üstələyir",
        "Gözləntiləri tam qarşılayır",
        "İnkişafa ehtiyac var",
        "Performans inkişaf planı (PIP)"
    ]
)

plt.show()
```

##### Bu qrafik menecerlər üzrə işçilərin performans nəticələrinin paylanmasını göstərir. Əksər menecerlərin komandasında əsas nəticə “Gözləntiləri 

##### tam qarşılayır” kimi görünür. Bəzi menecerlərdə isə “Gözləntiləri üstələyir” nəticəsi olan işçilərin sayı daha çoxdur. Bu da menecerlər arasında 

##### performans nəticələrində müəyyən fərqlərin olduğunu göstərir.

```python
plt.figure(figsize = [15, 5])
Ord = dataset['RecruitmentSource'].value_counts().index
sns.countplot(data= dataset, x='RecruitmentSource', hue ='PerformanceScore', order=Ord)
plt.xticks(rotation=45)
plt.title('İşə qəbul mənbəyinə görə performans')
plt.legend(
    title="Performans",
    labels=[
        "Gözləntiləri üstələyir",
        "Gözləntiləri tam qarşılayır",
        "İnkişafa ehtiyac var",
        "Performans inkişaf planı (PIP)"
    ]
)

plt.show()
```

##### Bu qrafik işçilərin hansı işə qəbul mənbəyindən gəldiyini və performans nəticələrinin paylanmasını göstərir. Görünür ki, ən çox işçi Indeed və 

##### LinkedIn vasitəsilə işə qəbul olunub. Bu mənbələrdə “Gözləntiləri tam qarşılayır” performans nəticəsi üstünlük təşkil edir. Digər mənbələrdə 

##### işçi sayı daha  azdır və nəticələr də nisbətən məhduddur.

```python
plt.figure(figsize = [15, 5])
Ord = dataset['RecruitmentSource'].value_counts().index
sns.countplot(data= dataset, x='RecruitmentSource', hue ='RaceDesc', order=Ord);
plt.xticks(rotation=45)
plt.legend(loc='upper right')
plt.title('İşə qəbul mənbəyi və irq bölgüsü');

plt.show()
```

##### Bu qrafik işə qəbul mənbələri üzrə əməkdaşların irq bölgüsünü göstərir. Görünür ki, əksər mənbələrdə işçilərin böyük hissəsi “White” 

##### kateqoriyasına aiddir. Bəzi mənbələrdə isə digər irq qruplarının payı nisbətən daha çoxdur
